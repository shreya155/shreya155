```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv ('C:\\Users\\shrey\\anaconda3\\Health.csv')
print(df.shape)
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
 
```

```python
from sklearn.impute import SimpleImputer
im = SimpleImputer(missing_values=np.nan,strategy='mean')
im
df.iloc[:,[1,2]]=im.fit_transform(df.iloc[:,[1,2]])
df.iloc[:,[1,2]]
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['Will survive till 70']= le.fit_transform(df['Will survive till 70'])
```

```python
le= LabelEncoder()
df['Ethnicity']= le.fit_transform(df['Ethnicity'])
df.head()
```

```python
i = df.iloc [:,:3]
d = df.iloc [:,3]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(i,d,test_size=.15)
print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
```

```python
xtrain.info()
```

```python
from sklearn.linear_model import LogisticRegression
LRclassifier=LogisticRegression()
LRclassifier.fit(xtrain,ytrain)
Prediction=LRclassifier.predict(xtest)
print("predicted y records:")
print(Prediction)
print(ytest)
print(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,Prediction))
```

```python
df= pd.read_csv('C:\\Users\\shrey\\anaconda3\\USA_housing.csv')
df.head()
```

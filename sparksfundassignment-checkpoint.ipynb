```python
df=pd.read_csv("C:\\Users\\shrey\\anaconda3\\rounds2.csv")
df1=pd.read_csv("C:\\Users\\shrey\\anaconda3\\mapping.csv")
df2=pd.read_csv("C:\\Users\\shrey\\anaconda3\\companies.csv")
df['raised_amount_usd']= pd.to_numeric(df['raised_amount_usd'],errors="coerce")
rest=df.groupby(["funding_round_type"])[['raised_amount_usd']].count()
print(rest)
```

```python
df9=df[df["funding_round_type"].str.contains('seed')==True]
print("average of seed type : :",df9["raised_amount_usd"].mean())
```

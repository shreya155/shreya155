```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df= pd.read_csv('C:\\Users\\shrey\\anaconda3\\fruit.csv')
df.head()
```

```python
i = df[['mass', 'width', 'height']]
d = df['fruit_label']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(i, d, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(xtrain,ytrain)
predict=knn.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predict)
```

```python
#wine assignment
df1= pd.read_csv('C:\\Users\\shrey\\anaconda3\\wine.csv')
df1.head()
```
